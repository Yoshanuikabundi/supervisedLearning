###### Logistic Regression

There are many linear models for classification. These algorithms all differ in the following two ways:
1. How they measure how well a particular combination of coefficients and intercept fits the training data.
2. If and what kind of regularization they use.

most common linear classification algorithm is Logistic Regression. Despite its name, LogisticRegression is a classification algorithm and not a regression algorithm, it is actually a probabilistic classification model

The formula for logistic regression is derived from the standard linear regression equation

$$ h(x) = w[0]x[0] + w[1]x[1] + .... + w[p]x[p] + b $$

$$ h(x) = w.x + b => w^Tx + b$$

In regression output is a linear function of features (a line/ a plane/ a hyperplane)
but in classification, decision boundary is a linear function of the input. in other words a binary/linear classifier is a classifier that separates two classes using a line, a plane or a hyperplane.

To predict the class of a binomial target feature, standard linear regression equation is transformed to logistic regression formula by wrapping it in a sigmoid function. And instead of returning the weighted sum of the features as in Linear regression, we set a threshold for the predicted values at zero.

$$ f(x) = \frac{1}{1+e^{-h(x)}} \begin{cases}+1 & h(x) >= 0\\0 & h(x) < 0\end{cases}$$

Here f(x) is the estimated probability that y=1 on given input x.
$$ f(x) = p(y=1|x;w) $$

$$ p(y=1|x;w) + p(y=0|x;w) == 1 $$

Hence $$ p(y=0|x;w) = 1 - p(y=1|x;w) $$

It's important to note that the data does not tell you the probability of a label but rather what label the sample has after being generated by the target distribution. Or we can say that Decision boundary is the property of the hypothesis we have considered, but not of the traning data. Training data is used to fit the parameters w and b.

The important assumptions of the logistic regression model include:

- Target variable is binary
- Predictive features are categorical or interval (if continuous)
- Features are independent of one another
- No imbalance in data


###### Binary classification

```python
# importing dependencies
from sklearn.linear_model import LogisticRegression
import mglearn
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# loading dataset
X, y = mglearn.datasets.make_forge()
print("X shape: {}".format(X.shape))
print("y values: {}".format(y[:5]))
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 3))
model, ax = [LogisticRegression(), axes]
clf = model.fit(X, y)
mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
ax.scatter(X[:, 0], X[:, 1], c=y, s=60, cmap=mglearn.cm2)
ax.set_title("%s" % clf.__class__.__name__)
```

We display the decision boundaries found by LogisticRegression as straight line, separating the area classified as red on the top from the area classified as blue on the bottom. However 2 points are misclassified and the model already had L2 regularizer applied by default (similar to ridge regression).

For LogisticRegression the trade-off parameter that determines the strength of the regularization is called C(sklearn package python), and higher values of C correspond to less regularization. In other words, when using a high value of the parameter C , Logistic Regression tries to fit the training set as best as possible, while with low values of the parameter C, the model put more emphasis on finding a coefficient vector w that is close to zero.

There is another interesting intuition of how the parameter C acts. Using low values of C will cause the algorithms try to adjust to the “majority” of data points, while using a higher value of C stresses the importance that each individual data point be classified correctly.

```python
# Considering Cancer Dataset for Logistic Regression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=42)
print(cancer.feature_names)
print(X_train.shape)
logisticregression = LogisticRegression().fit(X_train, y_train)
print("training set score: %f" % logisticregression.score(X_train, y_train))
print("test set score: %f" % logisticregression.score(X_test, y_test))
```

```python
# Increasin the value of c to 100
```

```python
logisticregression100 = LogisticRegression(C=100).fit(X_train, y_train)
print("training set score: %f" % logisticregression100.score(X_train, y_train))
print("test set score: %f" % logisticregression100.score(X_test, y_test))
```

```python
# Decreasing value of C to 0.01
logisticregression001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print("training set score: %f" % logisticregression001.score(X_train, y_train))
print("test set score: %f" % logisticregression001.score(X_test, y_test))
```

```python
# plotting all the coeff for various values of C
plt.plot(logisticregression.coef_.T, '^', label="C=1")
plt.plot(logisticregression100.coef_.T, 'o', label="C=100")
plt.plot(logisticregression001.coef_.T, '.', label="C=0.001")
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.ylim(-5, 5)
plt.legend()
```

Stronger regularization pushes coefficients more and more towards zero, though coefficients never become exactly zero, in the figure
for "mean perimeter", C=100 and C=1 the coeff is negative. While for C=0.001, the coeff is positive with greater magnitude. By this dataset indirectly people think that coeffecients tell us to which class a particular feature is associated with. But the output class label could be any one based on the model we consider among the available 3. Hence the change in the values of the coefficients should always be taken with utmost care.

```python
# using L1 regularization
for C in [0.001, 1, 100]:
    lr_l1 = LogisticRegression(C=C, penalty="l1").fit(X_train, y_train)
    print("training accuracy of L1 logreg with C=%f: %f"
    % (C, lr_l1.score(X_train, y_train)))
    print("test accuracy of L1 logreg with C=%f: %f"
    % (C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, 'o', label="C=%f" % C)
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
plt.ylim(-5, 5)
plt.legend(loc=2)
```

L1 regularizer gives a more interpretable model as most of the unimportant features are zeroed down, leaving only the are highly correlated with or impacting the target variable.

###### Pros & Cons
- Large alpha or small C mean simple models. Usually C and alpha are searched for on a logarithmic scale.
- Decide over the type of regularization you want to apply. Go with L1 regularization only if few features are unimportant.
- Models with L1 regularizer are easy to interpret as they use only few features. In these cases it is easy to explain the important features for the model.
- Linear models are very fast to train, and also fast to predict. They scale to very large datasets and work well with sparse data
- Unfortunately, it is often not entirely clear why coefficients are the way they are. This is particularly true if your dataset has highly correlated features, in those cases, the coefficients might be hard to interpret. As we cannot depict a features importance or contribution towards the target.
